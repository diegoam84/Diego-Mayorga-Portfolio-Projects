```python
text_reviews = []
for url in urls:
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html')
    reviews = soup.find_all('div', {'class':'text_content'})
    for review in reviews:
        to_append = review.text
        text_reviews.append(to_append)
```

```python
#Cheching if I got all the reviews from the webpage
len(text_reviews)
```

```python
dirty_data = text_reviews
```

```python
dirty_data[0]
```

```python
# Removing trip verified characters
first_cleaning = []
first_container = []
clean = []
for review in dirty_data:
    first_cleaning = review.replace('✅ Trip Verified |', '')
    first_container.append(first_cleaning)
    
for text in first_container:
    second_cleaning = text.replace('\xa0', '')
    clean.append(second_cleaning)
```

```python
# Gouping all reviews into one big text
reviews_combined = ''.join(clean)
```

```python
# Using RegEx characters to remove punctuation
import re
first_clean_combined = re.sub(r'[^\w\s)]', '', reviews_combined)
```

```python
second_clean = re.sub(r'[^\w\s)]', '', first_clean_combined)
```

```python
reviews_lower = second_clean.lower()
```

```python
# Splitting reviews into words to perform the count
reviews_broken_out = re.split(r'\s+', reviews_lower)
reviews_for_wordcloud = ' '.join(reviews_broken_out)
```

### Unsupervised Sentiment Analysis

```python
sentiment_text = []
for review in clean:
    sentiment_text.append(review.replace('Not Verified |', ''))
```

```python
sentiment_text[57]
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
# Create function to get the sentiment for each review
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    print("{:-<40} {}".format(sentence, str(score)))
```

Let's evaluate the performance of the sentiment analysis:

```python
sentiment_analyzer_scores(sentiment_text[0])
```

```python
sentiment_analyzer_scores(sentiment_text[3])
```

It succesfully identified a positive review and a bad review, now let's use it for all the reviews.

```python
scores = []
for review in sentiment_text:
    score = analyser.polarity_scores(review)
    scores.append(score)
```

```python
negative = []
neutral = []
positive = []
compound = []
for score in scores:
    negative.append(score['neg'])
    neutral.append(score['neu'])
    positive.append(score['pos'])
    compound.append(score['compound'])
```

```python
df = pd.DataFrame({'review':sentiment_text,
                   'negative':negative,
                  'neutral':neutral,
                  'positive':positive,
                  'compound':compound})
```

```python
df.head()
```

```python
df['sentiment'] = df.apply(lambda x: 'Positive' if x['compound'] >= 0 else 'Negative', axis=1)
```

Now that we managed to label our reviews as either positive or negative, we can do a new word cloud visualization but focusing on the good comments only, and then focusing on the bad comments only, in order to figure out what people like about British Airways, and what people don't like about British Airways.

```python
positive_reviews = list(df[df['sentiment']=='Positive']['review'])
negative_reviews = list(df[df['sentiment']=='Negative']['review'])
```

```python
def to_wordcloud(reviews):
    reviews_combined = ''.join(reviews) #joining all reviews into one big text
    first_clean_combined = re.sub(r'[^\w\s)]', '', reviews_combined) #remove punctuation
    second_clean = re.sub(r'[^\w\s)]', '', first_clean_combined) #second pass at removing punctuation
    to_plot = second_clean.lower() #making everything lowercase
    return to_plot
```

```python
positive_to_plot = to_wordcloud(positive_reviews)
```

```python
negative_to_plot = to_wordcloud(negative_reviews)
```

```python
positive_wordcloud = WordCloud(background_color='white', stopwords=review_stopwords, max_words=500).generate(positive_to_plot)
negative_wordcloud = WordCloud(background_color='white', stopwords=review_stopwords, max_words=500).generate(negative_to_plot)

fig, axs = plt.subplots(1,2, figsize=(20,15))

axs[0].imshow(positive_wordcloud, interpolation = 'bilinear')
axs[0].set_title('Most Mentioned Words in Positive Reviews')
axs[0].axis('off')

axs[1].imshow(negative_wordcloud, interpolation = 'bilinear')
axs[1].set_title('Most Mentioned Words in Negative Reviews')
axs[1].axis('off')

plt.show()
```

```python
positive_wordcloud.to_file('positive_wordcloud.png')
negative_wordcloud.to_file('negative_wordcloud.png')
```
