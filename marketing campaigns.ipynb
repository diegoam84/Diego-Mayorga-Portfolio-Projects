### Regression Analysis (Market Sales and Strategies)

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df.columns=[mystring.replace(' ', '_') for mystring in df.columns]
df.columns=[mystring.replace('(', '') for mystring in df.columns]
df.columns=[mystring.replace(')', '') for mystring in df.columns]
results = smf.ols('Amount_Collected ~ Campaign_Email+Campaign_Flyer+Campaign_Phone+Sales_Contact_1+Sales_Contact_2+Sales_Contact_3+Sales_Contact_4+Sales_Contact_5',data=df).fit()
print(results.summary())
```

```python
coef = pd.read_html(results.summary().tables[1].as_html(), header=0, index_col=0)[0]
coef
```

```python
coef = coef.reset_index()
coef = coef[coef['P>|t|']<0.05][['index', 'coef']]
coef
```

### Regression Analysis Broken up by Account Types

```python
summary = pd.DataFrame()
for acctype in list(set(list(df['Client_Type']))):
    print(acctype)
    temp = df[df['Client_Type']==acctype].copy()
    results = smf.ols('Amount_Collected ~ Campaign_Email+Campaign_Flyer+Campaign_Phone+Sales_Contact_1+Sales_Contact_2+Sales_Contact_3+Sales_Contact_4+Sales_Contact_5', data=temp).fit()
    coef = pd.read_html(results.summary().tables[1].as_html(), header=0, index_col=0)[0].reset_index()
    coef = coef[coef['P>|t|']<0.05][['index', 'coef']]
    coef.columns = ['Variable', 'Coefficient (Impact)']
    coef['Account Type'] = acctype
    coef = coef.sort_values('Coefficient (Impact)', ascending=False)
    coef = coef[coef['Variable']!= 'Intercept']
    summary = pd.concat([summary, coef])
    print(results.summary())
```

### Final Recommendations

Using the table below, we can use the coefficient to see how much return we can derive from each dollar we spend. Each account type has different methods that are effective to them.

```python
def format(x):
    return '${:.1f}'.format(x)
```

```python
summary.rename(columns={'Coefficient (Impact)':'Return On Investment'}, inplace=True)
summary['Return On Investment'] = summary['Return On Investment'].apply(lambda x: round(x,1))
```

```python
summary['Return On Investment'] = summary['Return On Investment'].apply(format)
```
